###  3. 대학 캠퍼스의 다양성 지수 발견

```python
college = pd.read_csv('data/college.csv')
college.head()
```

```python
college.info()
```

```python
college.isnull().sum().sum()
```

```python
college.isnull().sum()
```

```python
#특정 인덱스 넏기
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college.head()
```

```python
college_ug = college.filter(like='UGDS_')
college_ug.head()
```

```python
college_ug.isnull().sum(axis=0)
```

```python
(college_ug.isnull().sum(axis=0) > 0)
```

```python
(college_ug.isnull().sum(axis=0) > 0).sum()
```

```python
college_ug = college_ug.dropna(how='all')
```

```python
college_ug.isnull().sum()
```

```python
college_ug.head(3)
```

```python
(college_ug >= 0.15).sum(axis=0)
```

```python
(college_ug >= 0.15).sum(axis=1)
```

```python
college_metric = (college_ug >= 0.15).sum(axis=1)
college_metric.head()
```

```python
college_metric.value_counts()
```

```python
college_ug['UGDS_WHITE'].value_counts()
```

```python
#sorting
college_metric.sort_values(ascending=False)
```
