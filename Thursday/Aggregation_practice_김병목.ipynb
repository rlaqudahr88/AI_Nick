```python
import pandas as pd
import numpy as np
```

```python
reserve_tb = pd.read_csv('hotel_data/reserve.csv')
customer_tb = pd.read_csv('hotel_data/customer.csv')
hotel_tb = pd.read_csv('hotel_data/hotel.csv')
```

```python
reserve_tb
```

```python
# groupby로 hotel_id를 집약 단위로 설정하여,
# customer_id의 값을 nunique 함수를 이용하여 고객 수를 카운트

info = reserve_tb.groupby('hotel_id').agg({'customer_id':['nunique']})
level0 = info.columns.get_level_values(0)
level0
level1 = info.columns.get_level_values(1)
level1
info.columns = level0+'_'+level1
info.head(10)
```

```python
# agg 함수를 이용하여 집약 처리를 한꺼번에 지정
# reserve_id를 대상으로 count 함수를 적용
# customer_id를 대상으로 nunique 함수를 적용

reserve_tb.groupby('hotel_id').agg({'reserve_id':['count'],'customer_id':['nunique']})
# 딕션너리{}
#{index1: value1 }
```

```python
# 보너스 테스트용

reserve_tb.groupby('hotel_id')['reserve_id','customer_id'].agg(['count','nunique'])
```

```python
# 집약 단위를 hotel_id와 people_num의 조합으로 지정
# 집약한 데이터에서 total_price를 추출하여, sum함수에 적용하여 매출 합계 금액을 산출

sum_info = reserve_tb.groupby(['hotel_id','people_num']).agg({'total_price':['sum']})
level0 = sum_info.columns.get_level_values(0)
level0
level1 = sum_info.columns.get_level_values(1)
level1
sum_info.columns = level0+'_'+level1
sum_info.head(10)
```
